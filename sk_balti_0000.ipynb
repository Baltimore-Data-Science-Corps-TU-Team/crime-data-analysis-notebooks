```python
#import osmnx as ox
#import networkx as nx

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv('dat/BPD_Arrests.csv', low_memory=False)
```

```python
dat.info()
```

```python
race = dat.Race.value_counts()

fig = plt.figure(figsize=(6,4))
race.plot(kind='bar')

plt.title('Arrest by Race')
plt.ylabel('Total Cases')
plt.show()
```

```python
sex = dat.Sex.value_counts()

fig = plt.figure(figsize=(6,4))
sex.plot(kind='bar')

plt.title('Arrest by Sex')
plt.show()
```

```python
charge = dat.ChargeDescription.value_counts()
charge = charge[ charge.gt(2500) ]
labels = dat.ChargeDescription.array


fig = plt.figure(figsize=(6,4))
charge.plot(kind='bar')

plt.title("Most common charges")
plt.show()
```
